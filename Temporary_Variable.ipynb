# What is a temporary variable?

---



A temporary variable is simply a variable that exists to hold a value while a loop is running.

Think of it as a placeholder that changes its value each time through the loop.

**Example:** Understanding a Temporary Variable in a Loop
>



Suppose we have a dictionary that contains the following key-value pairs:

| Key        | Value      |
| ---------- | ---------- |
| `"No"`     | `"Ranjan"` |
| `"middle"` | `"Raghu"`  |
| `"Yes"`    | `"Segu"`   |

>
We also have a list containing the keys in the following order:

>

| Position | Value      |
| -------- | ---------- |
| 0        | `"No"`     |
| 1        | `"middle"` |
| 2        | `"Yes"`    |

>

The loop reads:

For each y in the list, retrieve d[y].

As the loop runs, the temporary variable y changes its value on each iteration.
>
| Iteration | Current Value of `y` | Python Substitutes | Result     |
| --------- | -------------------- | ------------------ | ---------- |
| 1         | `"No"`               | `d["No"]`          | `"Ranjan"` |
| 2         | `"middle"`           | `d["middle"]`      | `"Raghu"`  |
| 3         | `"Yes"`              | `d["Yes"]`         | `"Segu"`   |
>

The temporary variable acts like a placeholder. During the first iteration, it holds "No", so Python evaluates d["No"] and returns "Ranjan". During the second iteration, the value of y is replaced with "middle", so Python evaluates d["middle"] and returns "Raghu". During the third iteration, y holds "Yes", so Python evaluates d["Yes"] and returns "Segu".
>
The important point is that Python does not create a new variable for each iteration. It reuses the same variable (y) and simply updates its value each time through the loop.

This substitution technique works for dictionaries, lists, strings, tuples, and almost every loop you'll encounter in Python.

**Why Do We Use a Temporary Variable in a Loop?**

A temporary variable is used to hold the current item being processed in a loop.

As Python moves through a collection (such as a list, tuple, string, dictionary, or set), it processes one item at a time. The temporary variable provides a way to refer to that current item.

Without a temporary variable, Python would have no way of knowing which item you want to work with during each iteration.

**Rule to Remember**

Whenever you see:

for item in collection:

Read it as:

"Take each item from the collection, one at a time, store it temporarily in item, and use item while processing the current iteration."

**Where else are they used?**

| Construct                | Temporary Variable | Purpose                                               |
| ------------------------ | ------------------ | ----------------------------------------------------- |
| `for` loop               | `item`             | Holds the current item in the collection              |
| List comprehension       | `x`                | Holds each item while building a new list             |
| Dictionary comprehension | `k`, `v`           | Hold keys and values while building a dictionary      |
| Set comprehension        | `x`                | Holds each item while building a set                  |
| Generator expression     | `x`                | Holds each item while generating values one at a time |
| Lambda function          | `x`                | Holds the argument passed to the function             |
| Function parameter       | `name`             | Holds the argument while the function executes        |
| Exception handling       | `e`                | Holds the exception object temporarily                |


**Difference Between Normal, Temporary, and Parameter Variables**

| Feature                          | Normal Variable                             | Temporary Variable                                                                        | Parameter Variable                                                                 |
| -------------------------------- | ------------------------------------------- | ----------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------- |
| **Who creates it?**              | Programmer                                  | Python (during iteration or a construct like a comprehension), though you choose its name | Programmer defines it in a function; Python assigns it when the function is called |
| **Purpose**                      | Store data for general use                  | Hold the current value while Python performs a task                                       | Receive values (arguments) passed to a function                                    |
| **Lifetime**                     | Until it goes out of scope or is reassigned | Usually only while the loop/comprehension/task is executing                               | Only while the function is executing                                               |
| **Value changes automatically?** | ❌ No                                        | ✅ Yes (Python updates it each iteration)                                                  | ❌ No (set when the function is called)                                             |
| **Typical examples**             | `age`, `name`, `salary`                     | `item`, `x`, `key`, `value`                                                               | `name`, `age`, `x`, `y` in a function definition                                   |


**Rule to Remember**

| Variable Type          | Think of it as...                                                                                     |
| ---------------------- | ----------------------------------------------------------------------------------------------------- |
| **Normal Variable**    | A named storage location that **you manage**.                                                         |
| **Temporary Variable** | A placeholder that **Python updates automatically** while processing a sequence or similar construct. |
| **Parameter Variable** | A placeholder that **receives values passed into a function** and exists only for that function call. |
