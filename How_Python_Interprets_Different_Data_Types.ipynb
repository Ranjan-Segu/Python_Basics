# How Python Interprets Different Data Types

One of the most important ideas in Python is that every value has a type, and Python treats each type according to what it represents—not necessarily how humans see it.

---



| Data Type               | Example            | How Python Sees It                              | How Python **Doesn't** See It                     | Collection? |
| ----------------------- | ------------------ | ----------------------------------------------- | ------------------------------------------------- | ----------- |
| **Integer (`int`)**     | `123`              | One whole number                                | Three separate digits (`1`, `2`, `3`)             | ❌ No        |
| **Float (`float`)**     | `45.6`             | One floating-point number                       | Individual digits (`4`, `5`, `.`, `6`)            | ❌ No        |
| **Boolean (`bool`)**    | `True`             | One logical value (`True` or `False`)           | Multiple values or parts                          | ❌ No        |
| **String (`str`)**      | `"123"`            | A sequence of characters (`'1'`, `'2'`, `'3'`)  | One numeric value (123)                           | ✅ Yes       |
| **List (`list`)**       | `[10, 20, 30]`     | An ordered sequence of objects                  | One single object or number                       | ✅ Yes       |
| **Tuple (`tuple`)**     | `(10, 20, 30)`     | A fixed (immutable) ordered sequence of objects | One single object or number                       | ✅ Yes       |
| **Dictionary (`dict`)** | `{"a": 1, "b": 2}` | A mapping of **keys** to **values**             | An ordered sequence accessed by position*         | ✅ Yes       |
| **Set (`set`)**         | `{10, 20, 30}`     | An unordered collection of unique objects       | A sequence with a first, second, or third element | ✅ Yes       |


**Memory Tip**

Think of Python objects in two categories:

$$📦 Scalars (Single Values)$$

Python sees these as one value:

123 → One integer

45.6 → One floating-point number

True → One Boolean value

Python does not see them as collections of smaller parts.

$$Collections (Containers)$$

Python sees these as objects that contain other objects:

"Python" → Characters

[10, 20, 30] → Elements

(10, 20, 30) → Elements

{"a": 1, "b": 2} → Key-value pairs

{10, 20, 30} → Unique elements

The way you access those contained objects depends on the collection type:

Lists, tuples, and strings use indexes.
Dictionaries use keys.
Sets are unordered, so you can't directly access an element by index or key.

You might wonder, "But can't Python somehow access the digits of 123?"
