# Python [] (Subscript) Cheat Sheet: What Does object[variable] Mean?

always works in two steps:



1.   Python finds the value stored in the variable.

2.   Python interprets [] based on the type of object.



**Python Data Types and the Meaning of []**

---



| Object Type             | Example                             | Value in Variable | Python Evaluates To | Meaning of `[]`                      | Output                                           |
| ----------------------- | ----------------------------------- | ----------------- | ------------------- | ------------------------------------ | ------------------------------------------------ |
| **Dictionary (`dict`)** | `d = {"No":"Ranjan", "Yes":"Segu"}` | `y = "No"`        | `d["No"]`           | Look up by **key**                   | `"Ranjan"`                                       |
| **List (`list`)**       | `l = [10, 20, 30]`                  | `y = 1`           | `l[1]`              | Element at **index**                 | `20`                                             |
| **Tuple (`tuple`)**     | `t = (100, 200, 300)`               | `y = 2`           | `t[2]`              | Element at **index**                 | `300`                                            |
| **String (`str`)**      | `s = "Python"`                      | `y = 3`           | `s[3]`              | Character at **index**               | `"h"`                                            |
| **Set (`set`)**         | `st = {10, 20, 30}`                 | `y = 1`           | `st[1]`             | ❌ Not supported (sets are unordered) | `TypeError: 'set' object is not subscriptable`   |
| **Integer (`int`)**     | `n = 123`                           | `y = 0`           | `n[0]`              | ❌ Not supported                      | `TypeError: 'int' object is not subscriptable`   |
| **Float (`float`)**     | `f = 3.14`                          | `y = 1`           | `f[1]`              | ❌ Not supported                      | `TypeError: 'float' object is not subscriptable` |
| **Boolean (`bool`)**    | `b = True`                          | `y = 0`           | `b[0]`              | ❌ Not supported                      | `TypeError: 'bool' object is not subscriptable`  |
