# **How to ask for information about anything in python**

help()

Definition: This is a wrapper around pydoc.help that provides a helpful message
  when 'help' is typed at the Python interactive prompt.

 Calling help() at the Python prompt starts an interactive help session.
 Calling help(thing) prints help for the python object 'thing'.

```python
help(help)
```

dir()

Definition: Shows all the methods and attributes available for an object.

```python
help(dir)
```

**Common Terminologies used in python**

| Python Term           | Plain English                                                         | Example                                                                       |
| --------------------- | --------------------------------------------------------------------- | ----------------------------------------------------------------------------- |
| **Variable**          | A name that refers to an object                                       | `name = "Ranjan"` → `name` is the variable                                    |
| **Value**             | The actual data stored                                                | `"Ranjan"`, `25`, `[1, 2, 3]` are values                                      |
| **Object**            | A value that has data and behavior (methods/attributes)               | `"Ranjan"` is a string object, `[1,2,3]` is a list object                     |
| **Data Type**         | The category of an object                                             | `type("Ranjan")` → `<class 'str'>`                                            |
| **Function**          | Reusable code that you call                                           | `len("Python")` returns `6`                                                   |
| **Built-in Function** | A function provided by Python                                         | `print()`, `len()`, `type()`, `sorted()`                                      |
| **Method**            | A function that belongs to an object                                  | `"python".upper()` → `"PYTHON"`                                               |
| **Attribute**         | Information that belongs to an object (usually accessed without `()`) | `math.pi` → `3.141592653589793`                                               |
| **Keyword**           | A reserved word in Python's grammar                                   | `if`, `for`, `while`, `def`, `return`                                         |
| **Argument**          | The value passed into a function or method                            | In `len(name)`, `name` is the argument                                        |
| **Parameter**         | The variable in the function definition that receives the argument    | `def greet(name):` → `name` is the parameter                                  |
| **Return Value**      | The value a function gives back                                       | `len("Python")` returns `6`                                                   |
| **Module**            | A Python file containing functions, classes, and variables            | `import math` then `math.sqrt(16)`                                            |
| **Class**             | A blueprint for creating objects                                      | `class Student:`                                                              |
| **Instance**          | A specific object created from a class                                | `s = Student()` → `s` is an instance                                          |
| **Scope**             | The region where a name is visible                                    | A variable inside a function is usually only visible inside that function     |
| **Iterable**          | An object you can loop over                                           | `for ch in "Python":`, `for n in [1,2,3]:`                                    |
| **Mutable**           | Can be changed after creation                                         | `numbers = [1,2]`; `numbers.append(3)`                                        |
| **Immutable**         | Cannot be changed after creation                                      | `text = "abc"`; `text.upper()` returns a new string, it doesn't change `text` |
