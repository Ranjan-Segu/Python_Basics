| Keyword        | Purpose                | What Happens?                                                                             | Output (Using the Same Loop)   | Best Used When...                                                  |
| -------------- | ---------------------- | ----------------------------------------------------------------------------------------- | ------------------------------ | ------------------------------------------------------------------ |
| **`pass`**     | Placeholder            | Does nothing. Execution continues normally.                                               | `1 2 3 4 5`<br>`Loop Finished` | You need an empty block or haven't written the code yet.           |
| **`continue`** | Skip current iteration | Skips the remaining statements for the current iteration and moves to the next iteration. | `1 2 4 5`<br>`Loop Finished`   | You want to ignore certain items but continue processing the rest. |
| **`break`**    | Exit the loop          | Stops the loop immediately.                                                               | `1 2`<br>`Loop Finished`       | You have found what you need or want to stop the loop early.       |

**COMPARISON**

| Keyword        | What it Does                            | Effect on the Current Iteration              | Effect on the Loop                     |
| -------------- | --------------------------------------- | -------------------------------------------- | -------------------------------------- |
| **`pass`**     | Does nothing                            | Continues executing the remaining statements | Loop continues normally                |
| **`continue`** | Skips the rest of the current iteration | Stops the current iteration immediately      | Loop continues with the next iteration |
| **`break`**    | Exits the loop                          | Stops the current iteration                  | Loop ends immediately                  |


**Memory Trick**

Imagine you're reading a book.

pass = You pause for a second, then keep reading.

continue = You skip the rest of the current page and move to the next page.

break = You close the book completely.
